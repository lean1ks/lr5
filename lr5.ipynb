```python
def assign_income_category(df):
    if df['total_income'] in range(0,30001):
        return 'E'
    elif df['total_income'] in range(30001,50001):
        return 'D'
    elif df['total_income'] in range(50001,200001):
        return 'C'
    elif df['total_income'] in range(200001,1000001):
        return 'B'
    else:
        return 'A'
```

```python
df['total_income_category'] = df.apply(assign_income_category, axis=1)
df['purpose'].unique()
```

```python
def assign_purpose_category(df):
    auto_category = ["приобретение автомобиля","на покупку подержанного автомобиля","на покупку своего автомобиля","автомобили","сделка с подержанным автомобилем",
                     "автомобиль","свой автомобиль","сделка с автомобилем","на покупку автомобиля","операции с автомобилем"]
    real_estate_category =["покупка жилья","операции с жильем","покупка жилья для семьи","покупка недвижимости","покупка коммерческой недвижимости","покупка жилой недвижимости","строительство собственной недвижимости",
                          "недвижимость","строительство недвижимости","операции с коммерческой недвижимостью","строительство жилой недвижимости","покупка своего жилья","операции с недвижимостью","покупка жилья для сдачи",
                          "ремонт жилью"]
    wedding_category = ["сыграть свадьбу","на проведение свадьбы","свадьба"]
    if df['purpose'] in auto_category:
        return 'операции с автомобилем'
    elif df['purpose'] in real_estate_category:
        return 'операции с недвижимостью'
    elif df['purpose'] in wedding_category:
        return 'проведение свадьбы'
    else:
        return 'получение образования'
df['total_purpose_category'] = df.apply(assign_purpose_category, axis=1)
df
```
